```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img=cv2.imread('baby_image.jpg')
```

```python
img_rgb=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.figure(figsize=(20,8))
plt.imshow(img_rgb)
```

```python
lower = np.array([0, 48, 80], dtype = "uint8") #skin tone
upper = np.array([20, 255, 255], dtype = "uint8") #skin tone 
```

```python
img_hsv=cv2.cvtColor(img_rgb,cv2.COLOR_RGB2HSV)
```

```python
mask=cv2.inRange(img_hsv,lower,upper)
plt.figure(figsize=(20,8))
plt.imshow(mask)
```

```python
res=cv2.bitwise_and(img_rgb,img_rgb,mask=mask)
plt.figure(figsize=(20,8))
plt.imshow(res)
```

```python
res.shape
```

```python
r,g,b=res[100,100] #get a particular r g b value of pixel
```
