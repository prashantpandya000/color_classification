```python
import os 
import matplotlib.pyplot as plt
import numpy as np
from skimage.io import imread
from skimage.transform import resize
```

```python
# data path

target=[]
images=[]
flat_data=[]
DATADIR= 'D:\color_detection\color_classification_svm\images'

CATEGORIES= ['Beige Color','Canary Yellow color','lemon chiffon color','Yellow Chartreuse color','yellow color']
#label encoding the value 
for category in CATEGORIES:
    class_num=CATEGORIES.index(category)#labeling
    path=os.path.join(DATADIR,category)#path storing 
    print(path)
    for img in os.listdir(path):
        img_array=imread(os.path.join(path,img))
        
        img_resized=resize(img_array,(150,150,3))
        flat_data.append(img_resized.flatten())
        images.append(img_resized)
        target.append(class_num)

flat_data=np.array(flat_data)
target=np.array(target)
images=np.array(images)
```

```python
flat_data
```

```python
unique,count=np.unique(target,return_counts=True)
plt.bar(CATEGORIES,count)
```

```python
#split data into train and test 

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(flat_data,target,test_size=0.3,random_state=22)
```

```python
# SVM classification algorithm 
from sklearn.model_selection import GridSearchCV
from sklearn import svm

param_gird=[
    {'C':[1,10,100,1000],'kernel':['linear']},
    {
        'C':[1,10,100,1000],'gamma':[0.001,0.0001],'kernel':['rbf']
    },
]

svc=svm.SVC(probability=True)
clf=GridSearchCV(svc,param_gird)
clf.fit(x_train,y_train)
```

```python
y_pred=clf.predict(x_test)
y_pred # predicted output
y_test # actual output
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix

accuracy_score(y_pred,y_test)
```

```python
confusion_matrix(y_pred,y_test)
```

```python
# save model with pickle library
import pickle
pickle.dump(clf,open('img_model.p','wb'))
```
